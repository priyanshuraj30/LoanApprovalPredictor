```python
import numpy as np

class NaiveBayes:
    def __init__(self):
        self.class_probs = {}
        self.feature_probs = {}

    def fit(self, X_train, y_train):
        n_samples, n_features = X_train.shape
        self.classes = np.unique(y_train)
        
        # Calculate class probabilities
        for cls in self.classes:
            self.class_probs[cls] = np.sum(y_train == cls) / n_samples
            
            # Calculate mean and standard deviation for each feature per class
            class_data = X_train[y_train == cls]
            self.feature_probs[cls] = {
                "mean": np.mean(class_data, axis=0),
                "std": np.std(class_data, axis=0) + 1e-9  # Add small epsilon to avoid division by zero
            }

    def _gaussian_pdf(self, x, mean, std):
        exponent = np.exp(-((x - mean) ** 2) / (2 * (std ** 2)))
        return (1 / (np.sqrt(2 * np.pi) * std)) * exponent

    def predict(self, X_test):
        preds = []
        for x in X_test:
            class_scores = {}
            for cls in self.classes:
                class_prob = np.log(self.class_probs[cls])
                feature_probs = self.feature_probs[cls]
                for i, (mean, std) in enumerate(zip(feature_probs["mean"], feature_probs["std"])):
                    class_prob += np.log(self._gaussian_pdf(x[i], mean, std))
                class_scores[cls] = class_prob
            preds.append(max(class_scores, key=class_scores.get))
        return preds
```

```python
# Implementing the Principal Component analysis to decrease the dimension of the dataset
import numpy as np

class PCA:
    def __init__(self, n_components):
        self.n_components = n_components
        self.components = None

    def fit(self, X):
        # Compute mean of each feature
        self.mean = np.mean(X, axis=0)
        
        # Center the data
        X_centered = X - self.mean
        
        # Compute covariance matrix
        cov_matrix = np.cov(X_centered, rowvar=False)
        
        # Compute eigenvalues and eigenvectors of the covariance matrix
        eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
        
        # Sort eigenvalues and eigenvectors in descending order
        sorted_indices = np.argsort(eigenvalues)[::-1]
        sorted_eigenvalues = eigenvalues[sorted_indices]
        sorted_eigenvectors = eigenvectors[:, sorted_indices]
        
        # Convert n_components to integer
        if isinstance(self.n_components, float):
            total_variance = np.sum(sorted_eigenvalues)
            explained_variance = np.cumsum(sorted_eigenvalues) / total_variance
            num_components = np.argmax(explained_variance >= self.n_components) + 1
            self.n_components = num_components
        
        # Select the top n_components eigenvectors
        self.components = sorted_eigenvectors[:, :self.n_components]

    def transform(self, X):
        # Center the data
        X_centered = X - self.mean
        
        # Project the data onto the principal components
        return np.dot(X_centered, self.components)
```

```python
# Define the number of principal components
n_components = 0.95  # Retain 95% of the variance

# Apply PCA to the training data
pca = PCA(n_components)
pca.fit(x_train_resampled.values)
x_train_pca = pca.transform(x_train_resampled.values)

# Transform the test data using the same PCA object
x_test_pca = pca.transform(x_test.values)

# Initialize and fit the Naive Bayes model with PCA-transformed data
naive_bayes_model = NaiveBayes()
naive_bayes_model.fit(x_train_pca, y_train_resampled.values)

# Predict on the test set
predictions = naive_bayes_model.predict(x_test_pca)

# Evaluate the model
accuracy = np.mean(predictions == y_test.values)
print("Accuracy:", accuracy)
```
